```python
def plot_im(im, cm, do_plot = True):
    plt.figure(figsize = (15, 5))
    plt.imshow(im, vmin = 0, vmax = 1, cmap = cm)
    if do_plot : plt.show()
```

##  Training or Loading Pretrained NTM

## Train Model

```python
N = 128
M = 20
ntm = NeuralTM(N = N, M = M)
ntm.to(device)
```

```python
lr = 1e-4
momentum = 0.90
optimizer = torch.optim.RMSprop(ntm.parameters(), lr = lr, momentum=momentum)
criterion = torch.nn.BCELoss()
```

```python
def grad_clip(parameters, clip=10):
    for p in parameters:
        p.grad.data.clamp_(-clip, clip)
```

```python
losses = []
dl = data_loader(100000, 20, 1, 8, use_cuda = USE_CUDA)

PRINT_COUNTER = 1000
for ITERATION, (x, t) in enumerate(dl):
    optimizer.zero_grad()
    out, lists = ntm(x)
    loss = criterion(out, t)
    loss.backward()
    grad_clip(ntm.parameters(), 10)
    optimizer.step()
    losses += [float(loss.data.cpu().numpy())]
    if ITERATION % 100 == 0:
        print("Processing {}-th sequence.".format(ITERATION))
    if ITERATION % PRINT_COUNTER == 0:
        print("loss: ", np.mean(losses[-1000:]))
        f, (p1, p2, p3) = plt.subplots(1, 3)
        p1.imshow(x.data.cpu().numpy(),  vmin = 0, vmax = 1, cmap = cm)
        p2.imshow(t.data.cpu().numpy(),  vmin = 0, vmax = 1, cmap = cm)
        p3.imshow(out.data.cpu().numpy(),  vmin = 0, vmax = 1, cmap = cm)
        plt.show()
        f, (p1, p2) = plt.subplots(2, 1, figsize = (12, 4))
        p1.imshow(lists[0], vmin = 0.0, vmax = 1.0, cmap = cm)
        p2.imshow(lists[1], vmin = 0.0, vmax = 1.0, cmap = cm)
        f, (p1, p2) = plt.subplots(1, 2, figsize = (12, 4))
        p1.imshow(lists[2].T, vmin = 0.0, vmax = 1.0, cmap = cm)
        p2.imshow(lists[3].T, vmin = 0.0, vmax = 1.0, cmap = cm)
        plt.show()
        os.makedirs('checkpoints', exist_ok = True)
        torch.save(ntm, 'checkpoints/i-{}.pt'.format(ITERATION))
```

# Reproducing Generalization Results

```python
def transparent_cmap(cmap, N=255):
    "Copy colormap and set alpha values"

    mycmap = cmap
    mycmap._init()
    mycmap._lut[:,-1] = np.linspace(0, 0.8, N+4)
    return mycmap
cm_error = transparent_cmap(plt.cm.Reds)
```

```python
# if we increase the size, predictions becomes unstable, requiring several runs
for sz_seq in [10, 20, 60, 100]:
    print("Sequence size: {}".format(sz_seq))
    x, t = next(data_loader(1, sz_seq, sz_seq, 8, use_cuda = USE_CUDA))
    out, lists = ntm(x)
    f1, ((p1, p2)) = plt.subplots(2, 1, figsize = (10, 1.8))
    p1.imshow(t.data.cpu().numpy()[sz_seq+2:, :-2].T,  vmin = 0, vmax = 1, cmap = cm)
    p2.imshow(out.data.cpu().numpy()[sz_seq+2:, :-2].T,  vmin = 0, vmax = 1, cmap = cm)
    p2.imshow(t.data.cpu().numpy()[sz_seq+2:, :-2].T - out.data.cpu().numpy()[sz_seq+2:, :-2].T,  vmin = 0, vmax = 1, cmap = cm_error)
    plt.show()
```
